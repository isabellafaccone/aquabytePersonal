### Convert to 3d world coords

```python
import numpy as np


def euclidean_distance(p1, p2):
    return np.linalg.norm(np.array(p1) - np.array(p2))


def convert_to_world_point(x, y, d, parameters):
    """ from pixel coordinates to world coordinates """
    # get relevant parameters
    pixel_count_width = parameters["pixelCountWidth"]
    pixel_count_height = parameters["pixelCountHeight"]
    sensor_width = parameters["imageSensorWidth"]
    sensor_height = parameters["imageSensorHeight"]
    focal_length = parameters["focalLength"]

    image_center_x = pixel_count_width / 2.0
    image_center_y = pixel_count_height / 2.0
    px_x = x - image_center_x
    px_z = image_center_y - y

    sensor_x = px_x * (sensor_width / pixel_count_width)
    sensor_z = px_z * (sensor_height / pixel_count_height)

    # now move to world coordinates
    world_y = d
    world_x = (world_y * sensor_x) / focal_length
    world_z = (world_y * sensor_z) / focal_length
    out_coords = np.array([world_x, world_y, world_z])
    return out_coords


def depth_from_disp(disp, parameters):
    """ calculate the depth of the point based on the disparity value """
    focal_length_pixel = parameters["focalLengthPixel"]

    baseline = parameters["baseline"]
    depth = focal_length_pixel * baseline / np.array(disp)
    # if depth == np.inf or disp == -np.inf:
    #     print(disp, parameters)
    return depth


def pixel2world(left_crop, right_crop, parameters):
    """2D pixel coordinates to 3D world coordinates"""

    # first create a dic with crop keypoints
    image_coordinates = {"leftCrop": {},
                         "rightCrop": {}}
    for keypoint in left_crop:
        name = keypoint["keypointType"]
        image_coordinates["leftCrop"][name] = [keypoint["xFrame"], keypoint["yFrame"]]
    for keypoint in right_crop:
        name = keypoint["keypointType"]
        image_coordinates["rightCrop"][name] = [keypoint["xFrame"], keypoint["yFrame"]]

    # then loop through the right crop keypoints and calculate the world coordinates
    world_coordinates = {}
    for keypoint in left_crop:
        name = keypoint["keypointType"]
        disparity = image_coordinates["leftCrop"][name][0] - image_coordinates["rightCrop"][name][0]
        depth = depth_from_disp(disparity, parameters)
        world_point = convert_to_world_point(image_coordinates["leftCrop"][name][0],
                                             image_coordinates["leftCrop"][name][1],
                                             depth,
                                             parameters)
        world_coordinates[name] = world_point
    return world_coordinates

example_world_kps = pixel2world(example['annotation']['leftCrop'], example['annotation']['rightCrop'], example['camera_metadata'])
example_world_kps
```

```python
from tqdm import tqdm

tqdm.pandas()

def convert_keypoint_to_world(row):
    try:
        out_coords = pixel2world(
            row['annotation']['leftCrop'], 
            row['annotation']['rightCrop'], 
            row['camera_metadata'])
        all_coords = list(out_coords.values())
        # We get inf when the disparity is 0, which isn't supposed to happen, so we filter these out
        has_inf = np.inf in np.abs(np.concatenate(all_coords))
        if has_inf:
            return None
        return out_coords
    except:
        print(row)
        assert False

df['world_keypoints_coords'] = df.progress_apply(convert_keypoint_to_world, axis=1)
```

### Build relevant variables

```python
def get_center_x(kps):
    return np.median([kp[0] for kp in kps.values()])

def get_center_y(kps):
    return np.median([kp[2] for kp in kps.values()])

df['center_x'] = df['world_keypoints_coords'].progress_apply(get_center_x)
df['center_y'] = df['world_keypoints_coords'].progress_apply(get_center_y)
```

```python
def get_distance(kps):
    # median of all depths
    return np.median([kp[1] for kp in kps.values()])

df['distance_from_camera'] = df['world_keypoints_coords'].progress_apply(get_distance)
```

```python
def get_orientation(kps):
    slope_vec = (kps['TAIL_NOTCH'] - kps['UPPER_LIP'])
    norm_slope = slope_vec / np.linalg.norm(slope_vec)
    return norm_slope

df['orientation'] = df['world_keypoints_coords'].progress_apply(get_orientation)
```

```python
relevant_df.groupby('pen_id').aggregate('count')['url_key']
```

```python
bin_types = ['hour_bins', 'distance_bins', 'center_bins']

for col in bin_types:
    not_null = relevant_df[col].notnull()
    relevant_df = relevant_df[not_null]
    relevant_df[col] = pd.Categorical(relevant_df[col])
```

```python
relevant_df.groupby(['pen_id'] + bin_types).aggregate('count')
```

```python
TGT_STATES = [4, 6]
REFERENCE_STATES = [7]
TGT_SAMPLE_SIZE = 5
REFERENCE_SAMPLE_SIZE = 2

def get_comparisons(frame):
    is_tgt_pen = frame['pen_id'] == TARGET_PEN
    # is_tgt_state = frame['annotation_state_id'].isin(TGT_STATES)
    tgt_frame = frame[is_tgt_pen]# & is_tgt_state]
    tgt_frame['comparison_type'] = 'target'
    num_tgts = len(tgt_frame)
    if num_tgts:
        tgt_frame = tgt_frame.sample(min(TGT_SAMPLE_SIZE, num_tgts))
    else:
        return tgt_frame
    is_reference_pen = frame['pen_id'].isin(REFERENCE_PENS)
    #is_reference_state = frame['annotation_state_id'].isin(REFERENCE_STATES)
    reference_frame = frame[is_reference_pen]# & is_reference_state]
    reference_frame['comparison_type'] = 'reference'
    num_references = len(reference_frame)
    if num_references:
        reference_frame = reference_frame.sample(min(REFERENCE_SAMPLE_SIZE, num_references))
    else:
        return reference_frame
    return pd.concat([tgt_frame, reference_frame])

comparisons = relevant_df.groupby(bin_types).progress_apply(get_comparisons)
comparisons
```

```python
print('Num Doppelganger levels')
all_levels = list(set([row[:3] for row in comparisons.index]))
len(all_levels)
```

```python
import shutil

def download_file(url, local_path):
    response = req.get(url, stream=True)
    with open(local_path, 'wb') as out_file:
        shutil.copyfileobj(response.raw, out_file)
        
download_file(ex['left_crop_url'], ex['local_path'])
```

```python
ex_level = comparisons.loc[all_levels[4]]
ex_level
```

```python
import cv2 
import matplotlib.pyplot as plt

def process_level(level, pdf=None):
    references = []
    for pen_id in REFERENCE_PENS:
        this_pen = level[level['pen_id'] == pen_id]
        if len(this_pen):
            references.append(this_pen.iloc[0])
        
    tgt = level[level['comparison_type'] == 'target'].iloc[0]
    if tgt['local_path'] not in already_downloaded:
        download_file(tgt['left_crop_url'], tgt['local_path'])
    for ref in references:
        if ref['local_path'] not in already_downloaded:
            download_file(ref['left_crop_url'], ref['local_path'])
    title = make_title(tgt, references)
    make_comparison(tgt, references, title, pdf)
            
def make_title(tgt, refs):
    pen = f"Pen:{tgt['pen_id']}vs{','.join([str(ref['pen_id']) for ref in refs])}"
    short_data_bins = f"hour:{tgt['hour']},distance:{tgt['distance_bins']}"
    try:
        loc_bin = f"loc:{(round(tgt['center_bins'][0].left, 2), round(tgt['center_bins'][1].left,2))}"
    except:
        print('Warning:Failed to save center loc')
        loc_bin = ''
    return pen + short_data_bins + loc_bin
            
def make_comparison(tgt, refs, title, pdf):
    all_ims = [tgt] + refs
    dims = (0, 0)
    for idx, row in enumerate(all_ims):
        im = cv2.imread(row ['local_path'])
        im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB) 
        im_x, im_y = im.shape[0], im.shape[1]
        dims = (max(im_x, dims[0]), max(im_y, dims[1]))
        all_ims[idx] = im
        
    for idx, im in enumerate(all_ims):
        all_ims[idx] = cv2.copyMakeBorder(im, 0, dims[0] - im.shape[0], 0, dims[1] - im.shape[1], cv2.BORDER_CONSTANT)
    
    concatenated = np.concatenate(all_ims, axis=0)
    fig = plt.figure(figsize=(20, 20))
    plt.title(title, size=20)
    plt.imshow(concatenated)
    plt.axis('off')
    if pdf is not None:
        pdf.savefig(fig)
    
process_level(ex_level)
```

```python
from random import sample
from tqdm import tqdm
import matplotlib.backends.backend_pdf

def plot_comparisons(compare_bin_type, percategory_samples, pdf):
    this_bintype_idx = comparisons.index.names.index(compare_bin_type)
    levels = []
    for i, category in enumerate(comparisons[compare_bin_type].unique()):
        this_hour_levels = [lev for lev in all_levels if lev[this_bintype_idx] == category]
        levels += sample(this_hour_levels, min(percategory_samples, len(this_hour_levels)))
        
    for i, lev in tqdm(enumerate(levels)):
        this_level = comparisons.loc[lev]
        process_level(this_level, pdf)
    pdf.close()
```

```python
pdf = matplotlib.backends.backend_pdf.PdfPages("skorpo_hourly_comparisons.pdf")
plot_comparisons('hour_bins', 3, pdf)
```
